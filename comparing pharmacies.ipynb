```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time

# Function to scrape names and prices of medicines from a given website
def scrape_website(url):
    driver = webdriver.Chrome()  # You may need to adjust this path based on your system
    driver.get(url)
    time.sleep(3)  # Wait for the page to load

    # Enter search query for a medicine (assuming "paracetamol" as an example)
    search_box = driver.find_element_by_xpath("//input[@type='search']")
    search_box.send_keys("paracetamol")
    search_box.send_keys(Keys.RETURN)
    time.sleep(3)  # Wait for search results to load

    # Scrape names and prices of medicines
    medicines = {}
    try:
        # Scraping first 3 medicines (you can adjust this according to your requirement)
        for i in range(1, 4):
            name_xpath = f"//div[@class='itemName'][{i}]"  # XPath for medicine name
            price_xpath = f"//div[@class='itemPrice'][{i}]"  # XPath for medicine price
            name = driver.find_element_by_xpath(name_xpath).text
            price = driver.find_element_by_xpath(price_xpath).text
            medicines[name] = price
    except Exception as e:
        print(f"Error scraping: {str(e)}")

    driver.quit()
    return medicines

# Websites to scrape
websites = {
    "truemeds.in": "https://www.truemeds.in/",
    "pharmeasy.in": "https://www.pharmeasy.in/",
    "netmeds.com": "https://www.netmeds.com/",
    "1mg.com": "https://www.1mg.com/"
}

# Scrape each website and store the results
website_results = {}
for name, url in websites.items():
    website_results[name] = scrape_website(url)

# Find the cheapest medicine and its price across all websites
cheapest_medicine = None
cheapest_price = float('inf')

for website, medicines in website_results.items():
    for medicine, price in medicines.items():
        price = float(price.replace('₹', '').replace(',', ''))
        if price < cheapest_price:
            cheapest_price = price
            cheapest_medicine = (medicine, website)

print(f"The cheapest medicine is '{cheapest_medicine[0]}' from {cheapest_medicine[1]} priced at ₹{cheapest_price}")
```
